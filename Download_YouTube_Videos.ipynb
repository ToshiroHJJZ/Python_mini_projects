```python
import os
import warnings
warnings.filterwarnings('ignore')

from pytube import YouTube
from pytube.exceptions import AgeRestrictedError
from urllib.error import HTTPError

# Youtube video URL
url = 'https://www.youtube.com/watch?v=L7-c-LbNq9I&t=2485s'

try:
    # Create a Youtube object
    yt = YouTube(url)

    # Image file path (video path)
    video_path = "C:\\Users\\jonat\\Desktop\\Videos"

    # Get the highest resolution stream
    try:
        stream = yt.streams.get_highest_resolution()
    except AgeRestrictedError:
        print("The video is age-restricted. Trying to bypass age restriction...")
        stream = yt.streams.filter(progressive=True, file_extension='mp4').first()

    # Construct the full path (using video title as an example)
    video_title = yt.title
    full_path = os.path.join(video_path, f"{video_title}.mp4")

    # Download the video
    stream.download(output_path=video_path)

    # Output
    print(f"Successfully downloaded a YouTube video: {video_title}")
except HTTPError as e:
    print(f"HTTP Error: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

This script will download the video and audio separately in the best format available and then merge them into a single MP4 file. Please make sure you have ffmpeg installed on your system for this to work.

```python
import youtube_dl

ydl_opts = {
    'outtmpl': 'C:\\Users\\jonat\\Desktop\\Videos\\%(title)s.%(ext)s',
    'format': 'bestvideo+bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegVideoConvertor',
        'preferedformat': 'mp4',
    }],
}

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    info_dict = ydl.extract_info('https://www.youtube.com/watch?v=jyY3uRmKAZ4', download=True)
    video_title = info_dict.get('title', None)
    
print(f"Successfully downloaded a YouTube video: {video_title}")
```
